```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
countries = pd.read_csv('../datasets/countries-of-the-world.csv')
```

```python
gdp = countries['GDP ($ per capita)']
```

```python
phones = countries['Phones (per 1000)']
```

```python
literacy = countries['Literacy (%)']
```

```python
# Import Matplotlib and Seaborn
# import matplotlib.pyplot as plt
# import seaborn as sns

# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=gdp, y=phones)

# Show plot
plt.show()
```

```python
region = countries.iloc[0:11, 1]
```

```python
# Import Matplotlib and Seaborn
# import matplotlib.pyplot as plt
# import seaborn as sns


# Create count plot with region on the y-axis
sns.countplot(y=region)

# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
# import matplotlib.pyplot as plt
# import seaborn as sns

student_data = pd.read_csv('../datasets/student-alcohol-consumption.csv')

# Change the legend order in the scatter plot
sns.scatterplot(x="absences", y="G3", 
                data=student_data, 
                hue="location", hue_order=['Rural', 'Urban'])

# Show plot
plt.show()
```

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups

sns.countplot(x='school', hue = 'location',data=student_data, palette= palette_colors)


# Display plot
plt.show()
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            row="study_time")

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"], row = 'famsup', row_order = ['yes', 'no'])

# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
# import matplotlib.pyplot as plt
# import seaborn as sns
mpg = pd.read_csv('../datasets/mpg.csv')
# Create scatter plot of horsepower vs. mpg
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders", hue = 'cylinders')

# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
# import matplotlib.pyplot as plt
# import seaborn as sns

# Create a scatter plot of acceleration vs. mpg
sns.relplot(x='acceleration', y='mpg', data = mpg, hue = 'origin', style = 'origin', kind = 'scatter')



# Show plot
plt.show()
```

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci='sd')

# Show plot
plt.show()
```

```python
# Import Matplotlib and Seaborn
# import matplotlib.pyplot as plt
# import seaborn as sns

# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin", markers = True, dashes = False)

# Show plot
plt.show()
```

```python
# Create column subplots based on age category
sns.catplot(y="internet", data=student_data,
            kind="count", col = 'age')

# Show plot
plt.show()
```

```python
# Create a bar plot of interest in math, separated by gender
survey = pd.read_csv('../datasets/young-people-survey-responses.csv')
sns.catplot(x='Gender', y='Mathematics', data = survey, kind = 'bar')


# Show plot
plt.show()
```

```python
# Create bar plot of average final grade in each study category
order_categroy = ["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"]
sns.catplot(x='study_time', y = 'G3', data = student_data, kind = 'bar', order = order_categroy, ci = None)



# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])

# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci=None,
            estimator=median)

# Show plot
plt.show()
```
